```python
import numpy as np
import os
import time
from PIL import Image, ImageOps
from matplotlib import pyplot as plt
from IPython import display
import os
```

```python
PATH = "./mergedThingsOutlineOriginal/predictions/"
#indexes = [45, 3, 26, 2] dots2Outline
#indexes = [14,98,26, 50] Easy2Outline
indexes = []
```

```python
# using the variable axs for multiple Axes
fig, axs = plt.subplots(4, 3, figsize=(6,8))
fig.subplots_adjust(wspace=0, hspace=0.01)
axs[0][0].set_title("Input \n image", fontsize=10)
axs[0][1].set_title("Ground \n truth", fontsize=10)
axs[0][2].set_title("Predicted \n image", fontsize=10)

for i in range(len(axs)):
    axs[i][0].imshow(Image.open(f"{PATH}input_{indexes[i]}.jpg"))
    axs[i][1].imshow(Image.open(f"{PATH}target_{indexes[i]}.jpg"))
    axs[i][2].imshow(Image.open(f"{PATH}predictions_{indexes[i]}.jpg"))
    axs[i][0].axis('off')
    axs[i][1].axis('off')
    axs[i][2].axis('off')
```
